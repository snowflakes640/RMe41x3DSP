```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Generating sin and cos function 
angles = np.arange(0, 4*np.pi, .5)
sin_values = np.sin(angles)
plt.plot(angles, sin_values)
plt.show()
```

```python
cos_values = np.cos(angles)
plt.plot(angles, cos_values)
plt.show()
```

```python
# Generating a signal with noise
import random
num = []
t1 = []
for i in range(200):
    t1.append(i)
    num.append(random.randint(-20,20))

plt.plot(t1, num)
plt.show()
```

```python
# Unit impulse function
time = np.arange(-25,25,1)
value = []
for i in time:
    if i == 0: value.append(1)
    else: value.append(0)

plt.plot(time, value)
plt.show()
```

```python
# Unit step function
value = []
for i in time:
    if i<=5: value.append(0)
    else: value.append(1)

plt.stem(time, value)
plt.show()
```

```python
# Unit ramp function
time = []
value = []
for i in range(20):
    time.append(i)
    value.append(i)

plt.stem(time, value)
plt.show()
```

```python
# Exponential function
import math
time = np.arange(0, 700, .5)
value = pow(math.e, time)

plt.plot(time, value)
plt.show()
```
